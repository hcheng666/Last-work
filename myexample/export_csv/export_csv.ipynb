```python
import numpy as np
import pandas as pd

data = np.load('./data/data_avg.npy')
citys = pd.read_csv('./data/citys_coordiante.csv')
citys
```

```python
data_trans = []
city = citys['name']
for i in range(len(data)):
    for j in range(len(data)):
        if i==j:
            continue
        data_i = [0]*3
        data_i[0] = city[i]
        data_i[1] = city[j]
        data_i[2] = data[i][j]
        data_trans.append(data_i.copy())
pd.DataFrame(data_trans)
```

```python
edge_list = pd.DataFrame(data_trans)
edge_list.to_csv('./data/edge_list.csv',index=None)
```
